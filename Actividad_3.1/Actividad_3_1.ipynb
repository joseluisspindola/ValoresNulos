```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Ventas_totales.csv')
data.head(5)
```

```python
#Ver que columnas tienen valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizar una copia del archivo
data2=data.copy()
```

```python
#Valor en concreto - salón ventas
data2["salon_ventas"]= data2["salon_ventas"].fillna(2.756931e+07)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Media - tarjetas de débito
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data["tarjetas_debito"].mean(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Forward fill - tarjetas de crédito
data2["tarjetas_credito"] =data2["tarjetas_credito"].fillna(method="ffill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Backward fill  - otros medios
data2["otros_medios"] =data2["otros_medios"].fillna(method="bfill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Mediana - subotal ventas de bebidas y alimentos
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].median(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Media - bebidas
data2["bebidas"]=data2["bebidas"].fillna(round(data["bebidas"].mean(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valor en concreto - Almacén
data2["almacen"]= data2["almacen"].fillna(5.92e+06)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Forward fill - panaderia
data2["panaderia"] =data2["panaderia"].fillna(method="ffill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Mediana - lacteos
data2["lacteos"]=data2["lacteos"].fillna(round(data["lacteos"].median(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Backward fill  - otros medios
data2["carnes"] =data2["carnes"].fillna(method="bfill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Media - Verduleria y frutería
data2["verduleria_fruteria"]=data2["verduleria_fruteria"].fillna(round(data["verduleria_fruteria"].mean(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Forward fill - Rostisería
data2["alimentos_preparados_rotiseria"] =data2["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Backward fill  - Indumentaria
data2["indumentaria_calzado_textiles_hogar"] =data2["indumentaria_calzado_textiles_hogar"].fillna(method="bfill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Mediana - lacteos
data2["electronicos_articulos_hogar"]=data2["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Media - otros
data2["otros"]=data2["otros"].fillna(round(data["otros"].mean(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Descargar
data2.to_csv("ventas_totales_sin_nulos.csv")
```
