```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('JerseyCity.csv')
# Ajustar maximo de filas
pd.options.display.max_rows = None
data.head(5)
```

```python
#Ver que columnas tienen valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Eliminar columnas ya que no tiene registros
data1= data.drop([ 'calendar_updated'], axis=1)
data1
```

```python
#Valor en concreto - host location
data1["host_location"] = data1["host_location"].fillna("Sin ubicación del anfitrión")
data1.head()
```

```python
#Valor en concreto - response time
data1["host_response_time"] = data1["host_response_time"].fillna("Sin tiempo de respuesta promedio")
data1.head()
```

```python
#Media - host response rate
data1['host_response_rate']=data1['host_response_rate'].str.replace('%', '')
data1['host_response_rate']= data1['host_response_rate'].astype(float)
data1["host_response_rate"]=data1["host_response_rate"].fillna(round(data1["host_response_rate"].mean(),1))
```

```python
#Media - host acceptance rate
data1['host_acceptance_rate']=data1['host_acceptance_rate'].str.replace('%', '')
data1['host_acceptance_rate']= data1['host_acceptance_rate'].astype(float)
data1["host_acceptance_rate"]=data1["host_acceptance_rate"].fillna(round(data1["host_acceptance_rate"].mean(),1))
```

```python
#Valor en concreto - host is superhost
data1["host_is_superhost"] = data1["host_is_superhost"].fillna("Sin información")
data1.head()
```

```python
#Valor en concreto - host neighbourhood
data1["host_neighbourhood"] = data1["host_neighbourhood"].fillna("Sin información del vecindario")
data1.head()
```

```python
#Valor en concreto - bathrooms
data1["bathrooms"] = data1["bathrooms"].fillna("50")
data1.head()
```

```python
#Valor en concreto - bedrooms
data1["bedrooms"] = data1["bedrooms"].fillna("50")
```

```python
#Valor en concreto - beds
data1["beds"] = data1["beds"].fillna("50")
```

```python
#Media - price
data1['price']=data1['price'].str.replace('$', '')
data1['price'] = data1['price'].str.replace(',', '')
data1['price']= data1['price'].astype(float)
data1["price"]=data1["price"].fillna(round(data1["price"].mean(),1))
```

```python
#Valor en concreto - has avaliability
data1["has_availability"] = data1["has_availability"].fillna("Sin informacion")
data1.head()
```

```python
#Valor en concreto - estimated_revenue_l365d
data1["estimated_revenue_l365d"] = data1["estimated_revenue_l365d"].fillna("500000")
data1.head()
```

```python
#Valor en concreto - first y last review
data1[["first_review","last_review"]] = data1[["first_review","last_review"]].fillna("No hay fechas")
data1.head()
```

```python
#Valor en concreto - scores
data1[["review_scores_rating","review_scores_accuracy","review_scores_cleanliness", "review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value"]] = data1[["review_scores_rating","review_scores_accuracy","review_scores_cleanliness", "review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value"]].fillna("0")
data1.head()
```

```python
#Valor en concreto - reviews per month
data1["reviews_per_month"] = data1["reviews_per_month"].fillna("50")
data1.head()
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Descargar
data1.to_csv("JerseyCitySinNulos.csv")
```
